```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,Flatten
from tensorflow.keras.datasets import mnist
```

```python
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
model=Sequential()

model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=(28,28,1)))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))
model.add(Conv2D(32,kernel_size=(3,3),padding='valid',activation='relu'))

model.add(Flatten())

model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model1=Sequential()
model1.add(Conv2D(32,kernel_size=(3,3),padding='same',activation='relu',input_shape=(28,28,1)))
model1.add(Conv2D(32,kernel_size=(3,3),activation='relu',padding='same'))
model1.add(Conv2D(32,kernel_size=(3,3),activation='relu',padding='same'))

model1.add(Flatten())
model1.add(Dense(128,activation='relu'))
model1.add(Dense(10,activation='softmax'))
```

```python
model1.summary()
```

```python
# stride
```

```python
model2=Sequential()
model2.add(Conv2D(32,activation='relu',input_shape=(28,28,1),strides=(2,2),kernel_size=(3,3)))
model2.add(Conv2D(32,activation='relu',strides=(2,2),kernel_size=(3,3)))
model2.add(Conv2D(32,activation='relu',strides=(2,2),kernel_size=(3,3)))

model2.add(Flatten())

model2.add(Dense(128,activation='relu'))
model2.add(Dense(10,activation='softmax'))

model2.summary()
```

```python
# stride + Padding
```

```python
model3=Sequential()
model3.add(Conv2D(32,activation='relu',padding='same',kernel_size=(3,3),strides=(2,2),input_shape=(28,28,1)))
model3.add(Conv2D(32,activation='relu',padding='same',kernel_size=(3,3),strides=(2,2)))
model3.add(Conv2D(32,activation='relu',padding='same',kernel_size=(3,3),strides=(2,2)))

model3.add(Flatten())

model3.add(Dense(128,activation='relu'))
model3.add(Dense(10,activation='softmax'))
```

```python
model3.summary()
```
