```python
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from mlxtend.plotting import plot_decision_regions
import numpy as np
```

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/jay-kanakia/Deep-Learning-CampusX/refs/heads/main/Datasets/ushape.csv',names=['X1',"X2",'y'])
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
def random_init(act):
  model=Sequential()
  model.add(Dense(10,activation=act,input_dim=X.shape[1]))
  model.add(Dense(1,activation='sigmoid'))
  print(model.summary())
  initial_weight=model.get_weights()

  initial_weight[0]=np.random.sample(initial_weight[0].shape)*0.0001
  initial_weight[1]=np.random.sample(initial_weight[1].shape)*0.0001
  initial_weight[2]=np.random.sample(initial_weight[2].shape)*0.0001
  initial_weight[3]=np.random.sample(initial_weight[3].shape)*0.0001


  model.set_weights(initial_weight)
  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  history=model.fit(X,y,epochs=250,verbose=False,validation_split=0.2)

  plt.figure(figsize=(12,12))
  plt.subplot(2,2,1)
  sns.lineplot(history.history['accuracy'])
  sns.lineplot(history.history['val_accuracy'])

  plt.subplot(2,2,2)
  sns.lineplot(history.history['loss'])
  sns.lineplot(history.history['val_loss'])

  print()
  print('*'*100)
  print()

  print('new model weights are\n\n',model.get_weights())
  print()
  print('*'*100)
  print()
  plt.subplot(2,2,3)
  return plot_decision_regions(X.to_numpy(),y.astype('int').to_numpy(),clf=model)

```

```python
random_init('sigmoid')
```

```python
random_init('tanh')
```
