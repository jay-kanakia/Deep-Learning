```python
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from mlxtend.plotting import plot_decision_regions
import numpy as np
```

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/jay-kanakia/Deep-Learning-CampusX/refs/heads/main/Datasets/ushape.csv',names=['X1',"X2",'y'])

```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
def init_tech(act,ker_init):
  model=Sequential()
  model.add(Dense(10,activation=act,input_dim=X.shape[1],kernel_initializer=ker_init))
  model.add(Dense(10,activation=act,kernel_initializer=ker_init))
  model.add(Dense(10,activation=act,kernel_initializer=ker_init))
  model.add(Dense(10,activation=act,kernel_initializer=ker_init))
  model.add(Dense(1,activation='sigmoid'))
  print(model.summary())

  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  history=model.fit(X,y,epochs=100,verbose=False,validation_split=0.2)

  plt.figure(figsize=(8,8))
  plt.subplot(2,2,1)
  sns.lineplot(history.history['accuracy'])
  sns.lineplot(history.history['val_accuracy'])

  plt.subplot(2,2,2)
  sns.lineplot(history.history['loss'])
  sns.lineplot(history.history['val_loss'])

  print()
  print('*'*100)
  print()

  print('new model weights are\n\n',model.get_weights())
  print()
  print('*'*100)
  print()
  plt.subplot(2,2,3)
  return plot_decision_regions(X.to_numpy(),y.astype('int').to_numpy(),clf=model)
```

```python
init_tech('tanh','GlorotNormal')
```

```python
init_tech('relu','HeUniform')
```

```python
init_tech('relu','HeNormal')
```

```python
init_tech('sigmoid','GlorotUniform')
```
