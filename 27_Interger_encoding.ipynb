```python
import numpy as np

docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
token=Tokenizer()
```

```python
token.fit_on_texts(docs)
```

```python
token.word_counts
```

```python
token.document_count
```

```python
token.word_index
```

```python
seq=token.texts_to_sequences(docs)
seq
```

```python
seq=pad_sequences(seq,padding='post')
```

```python
seq
```

```python
from tensorflow.keras.layers import Embedding,Dense,SimpleRNN
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
model.add(Embedding(18,output_dim=3,input_length=5))
model.add(SimpleRNN(3,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile('adam','accuracy')
```

```python
model.predict(seq)
```

```python
from tensorflow.keras.datasets import imdb
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data()
```

```python
token=Tokenizer(oov_token='OOV')
X_train[0]
```

```python
len(max(X_train))
```

```python
X_train_padded=pad_sequences(X_train,padding='post')
X_test_padded=pad_sequences(X_train,padding='post')
```

```python
len(X_train_padded[55])
```

```python
X_train.shape
```

```python
X_train_padded.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN,Dense,Input
```

```python
from keras import activations
model=Sequential()
model.add(Input(shape=(2494,)))     # define sequence length
model.add(Embedding(input_dim=90000, output_dim=2))
model.add(SimpleRNN(32,activation='relu',return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train_padded,y_train,epochs=2,validation_data=(X_test_padded,y_test))
```
