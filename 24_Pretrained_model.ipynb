```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import ResNet50,preprocess_input,decode_predictions
```

```python
import numpy as np
```

```python
model=ResNet50(weights='imagenet')
```

```python
img_path='/content/burger.jpg'
img=image.load_img(img_path,target_size=(224,224))
img
```

```python
x=image.img_to_array(img)
```

```python
x.ndim
```

```python
y=np.expand_dims(x,axis=0)
```

```python
y.ndim
```

```python
y=preprocess_input(y)
```

```python
predicted_array=model.predict(y)
```

```python
decode_predictions(predicted_array)[0]
```

```python
import matplotlib.pyplot as plt
```

```python
def predict_image(image_path):
  img=image.load_img(image_path,target_size=(224,224))
  x=image.img_to_array(img)
  plt.imshow(img)
  x=np.expand_dims(x,axis=0)
  x=preprocess_input(x)
  pred=model.predict(x)
  return decode_predictions(pred)[0]
```

```python
burger_path='/content/burger.jpg'
```

```python
headphone_path='/content/headphone.jpg'
```

```python
speaker_path='/content/speaker.jpg'
```

```python
predict_image(burger_path)
```

```python
predict_image(headphone_path)
```

```python
predict_image(speaker_path)
```
